# Adding a column to mention Normal/Abnormaral blood pressure.

```python
df["bp_cat"] = df.BloodPressure.apply(lambda x: "Normal" if x in range(60,101) else "Low" if x<60 else "High")
df.head()
```

```python
data = {'Employee id': [10, 20, 15, 25, 30],
        'Gender': ['M', 'F', 'F', 'M', 'F'],
        'Remarks': ['Good', 'Nice', 'Good', 'Great', 'Nice'],
        }
df1 = pd.DataFrame(data)
print(f"Employee data : \n{df1}")

categorical_columns = df1.select_dtypes(include=['object']).columns.tolist()
encoder = OneHotEncoder(sparse_output=False)

one_hot_encoded = encoder.fit_transform(df1[categorical_columns])

one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(categorical_columns))

df_encoded = pd.concat([df1, one_hot_df], axis=1)

df_encoded = df_encoded.drop(categorical_columns, axis=1)
print(f"Encoded Employee data : \n{df_encoded}")
```

```python
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
enc = encoder.fit_transform(df[['bp_cat']])
enc_df = pd.DataFrame(enc, columns=encoder.get_feature_names_out(['bp_cat']))
enc_df.head()
```

```python
one_hot_df = pd.concat([df.drop(columns=['bp_cat','BloodPressure']), enc_df], axis=1)
one_hot_df.head()
```

```python
train_data = one_hot_df.sample(frac=0.8,random_state=1).reset_index(drop=True)
test_data = one_hot_df.drop(index=train_data.index).reset_index(drop=True)
print("Data for modelling - Training "+str(train_data.shape))
print("Data for prediction - Testing"+str(test_data.shape))
```

```python
classification_model = setup(data=train_data,target="Outcome",session_id=2143)
```
