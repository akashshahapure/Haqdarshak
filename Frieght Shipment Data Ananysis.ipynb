# Cleaning the data

```python
cleaned_data = data.dropna(subset=['Weight_KG','Shipping_Date']).copy()
cleaned_data.info()
```

```python
cleaned_data.Incoterms.fillna('Unknown', inplace=True)
cleaned_data.Status.fillna('B_In Transit', inplace=True)
cleaned_data.Sales_Channel.fillna('Unknown', inplace=True)
cleaned_data.info()
```
