# Checking outlier

```python
sns.scatterplot(txnData[['PROD_QTY', 'TOT_SALES']])
plt.show()
```

```python
plt.boxplot(txnData[['PROD_QTY', 'TOT_SALES']], labels=['PROD_QTY', 'TOT_SALES'], manage_ticks=True)
plt.show()
```

### Findings
- As per above investigation 'PROD_QTY' & 'TOT_SALES' are the columns 
which can be used for further analysis and above graph shows that
PROD_QTY' has outlier whihc is greater than 200 and 'TOT_SALES'
has outlier greater than 600.
- Also we can see there is relation between QTY vs Sales as if we
filterout data by excluding 'PROD_QTY' > 100 then we can deal with
extreme outliers.

```python
plt.boxplot(txnData[(txnData.PROD_QTY < 100)][['PROD_QTY', 'TOT_SALES']], labels=['PROD_QTY', 'TOT_SALES'], manage_ticks=True)
plt.show()
```

```python
plt.scatter(data = txnData[txnData.PROD_QTY < 100], x = 'TOT_SALES', y = 'PROD_QTY')
plt.legend(['TOT_SALES', 'PROD_QTY'])
plt.show()
```

```python
noOtlrData = txnData[txnData.PROD_QTY < 100]
noOtlrData.info()
```

```python
noOtlrData.describe()
```

```python
pName = []
weight = []

for prod in noOtlrData.PROD_NAME:
    for i in range(len(prod)):
        if prod[i].isnumeric():
            pName.append(' '.join(prod[0:i].split()))
            wt = prod[i:]
            for j in range(len(wt)):
                if len(wt) < 5:
                    weight.append(wt)
                    break
                elif wt[j] == 'g':
                    weight.append(wt[0:j+1])
                    break
            break

noOtlrData.PROD_NAME = pName
noOtlrData['PROD_WT'] = weight

noOtlrData.head(10)
```

# Lets check correlation between selected columns

```python
cols = noOtlrData.select_dtypes(exclude=['datetime64', 'object']).columns.to_list()
cols
```

```python
sns.heatmap(noOtlrData[cols].corr())
plt.show()
```

- As per above graph found below insights:
  1. STORE_NBR & TXN_ID is highly correlated with LYLTY_CARD_NBR
  2. As we saw in scatterplot graph, above graph shows that PROD_QTY & TOT_SALES are also corelated.


- Now, lets merge transaction data with purchase behavior data but first check the purhase behavior data

```python
purBehData.info()
```

```python
purBehData.LIFESTAGE.value_counts()
```

```python
purBehData.PREMIUM_CUSTOMER.value_counts()
```

```python
mergeData = noOtlrData.merge(purBehData, how='left', on='LYLTY_CARD_NBR')
mergeData.head()
```

```python
mergeData.info()
```

# Total Sales

```python
print('Total Sales = $ {0}/-'.format(int(mergeData.TOT_SALES.sum())))
```

# Top 10 products by sales value

```python
top10ProdSale = pd.DataFrame(mergeData.groupby(by = 'PROD_NAME').TOT_SALES.sum()).reset_index(drop=False).sort_values(by='TOT_SALES', ascending=False).head(10)
sns.barplot(data = top10ProdSale, x = 'TOT_SALES', y = 'PROD_NAME', hue = 'PROD_NAME')
plt.show()
```

```python
wtWiseSale = pd.DataFrame(mergeData.groupby(by='PROD_WT').TOT_SALES.sum()).reset_index(drop=False).sort_values(by='TOT_SALES', ascending=False)
ax = sns.barplot(data = wtWiseSale.head(10), x = 'TOT_SALES', y = 'PROD_WT')
ax.bar_label(ax.containers[0], fontsize = 8)
plt.show()
```

```python
premSales = pd.DataFrame(mergeData.groupby(by='PREMIUM_CUSTOMER').TOT_SALES.sum()).reset_index(drop=False).sort_values(by='TOT_SALES', ascending=False)
ax = sns.barplot(data = premSales, x = 'PREMIUM_CUSTOMER', y = 'TOT_SALES', )
ax.bar_label(ax.containers[0], fontsize=10)
plt.show()
```

```python
mergeData.info()
```

```python
lifestageSales = pd.DataFrame(mergeData.groupby(by='LIFESTAGE').TOT_SALES.sum()).reset_index(drop=False).sort_values(by='TOT_SALES', ascending=False)
ax = sns.barplot(lifestageSales, y='LIFESTAGE', x='TOT_SALES')
ax.bar_label(ax.containers[0], fontsize=8)
plt.show()
```
